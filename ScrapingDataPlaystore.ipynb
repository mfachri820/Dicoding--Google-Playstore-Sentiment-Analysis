```python
from google_play_scraper import Sort, reviews
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
# Scrape reviews
result, continuation_token = reviews(
    'com.gojek.app',
    lang='id',  # language
    country='id',  # country
    sort=Sort.MOST_RELEVANT,  # sorting method
    count=3000,  # number of reviews
    filter_score_with=None  # filter by score
)

# Convert the results to a DataFrame
data = pd.DataFrame(result)
data.head()
```

```python
# Dataframe dengan nama
data = pd.DataFrame(np.array(result),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head()
```

```python
len(data)
```

```python
data = data[['content','score']]
data.head()
```

```python
data = data.rename(columns={'content': 'ulasan', 'score': 'label'}) #mengganti nama fitur
data.head()
```

```python
data.to_csv("Ulasan Gojek 3000 Data.csv", index = False  , encoding='utf-8')
```
